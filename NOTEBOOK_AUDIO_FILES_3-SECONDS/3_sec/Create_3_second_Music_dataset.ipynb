```
import os
from pydub import AudioSegment

path = "/content/Data/genres_original/blues"

os.chdir(path)

audio_files = os.listdir()

for file in audio_files:

    name, ext = os.path.splitext(file)
    if ext == ".wav":
        w = AudioSegment.from_wav(file)
        chunk_length_ms = 3000 
        chunks = make_chunks(w, chunk_length_ms) 
        for i, chunk in enumerate(chunks):
            chunk_name = "{0}{1}.wav".format(name, chr(ord('a')+i))
            print ("exporting", chunk_name)
            chunk.export('/content/drive/MyDrive/Colab Notebooks/PROGETTO MUSICA DSIM/DATASET/DATASET_3_SEC/blues/' + chunk_name, format="wav")
```
