```
import librosa 
import numpy as np
import pandas as pd
import os
from matplotlib import pyplot as plt
import librosa.display
import pylab
from matplotlib import cm
```

```
n_fft=2048
hop_length=512
IMG_DIR='/content/MEL_IMAGES/'

for f in sorted(os.listdir('/content/Data/genres_original/blues/')):
        if f.endswith('.wav'):

            # Load file and compute the requested features
            y, sr = librosa.load('/content/Data/genres_original/blues/' + f)
            y_, sr_ = librosa.effects.trim(y)

            # Mel Spectogram
            mel_blues_= librosa.feature.melspectrogram(y_, sr=sr, n_fft=n_fft, hop_length=hop_length)
            mel_blues = librosa.amplitude_to_db(mel_blues_, ref=np.max)
        
            # Plotting the spectrogram and save as JPG without axes (just the image)
            pylab.figure(figsize=(3,3))
            pylab.axis('off') 
            pylab.axes([0., 0., 1., 1.], frameon=False, xticks=[], yticks=[]) # Remove the white edge
            librosa.display.specshow(mel_blues, cmap='cool')
            pylab.savefig(IMG_DIR + f[:-4]+'.jpg', bbox_inches=None, pad_inches=0)
            pylab.close()
```
