## Blues

### Audio features

```
%%time

# Audio features blues

SR_list_blues = []
CENTROID_blues = []
BANDWIDTH_blues = []
ROLLOFF_blues = []
ZCR_blues = []
HARMONIC_blues = []
PERCUSSIVE_blues = []
BMP_blues = []
RMSE_blues = []
CHROMA_blues=[]
MFCC_blues = []
SR_list = []

for f in sorted(os.listdir('/content/Data/genres_original/blues/')):
        if f.endswith('.wav'):

            # Load file and compute the requested features
            y, sr = librosa.load('/content/Data/genres_original/blues/' + f)
            SR_list_blues.append(sr)
            
            # ZCR
            zcr_blue = librosa.feature.zero_crossing_rate(y)
            ZCR_blues.append(zcr_blue)

            # Harmonic & Perceptrual
            y_harmonic_blue, y_percussive_blue = librosa.effects.hpss(y)
            HARMONIC_blues.append(y_harmonic_blue)
            PERCUSSIVE_blues.append(y_percussive_blue)
            
            # BPM
            tempo_blues, beat_times_blues = librosa.beat.beat_track(y, sr=sr, units='time')
            BMP_blues.append(tempo_blues)

            # RMSE
            rmse_blues = librosa.feature.rmse(y, center=True)
            RMSE_blues.append(rmse_blues)

            # CHROMA features
            chroma_blues = librosa.feature.chroma_stft(y, sr=sr, hop_length= 3000)
            CHROMA_blues.append(chroma_blues)

            # MFCC
            mfcc_blues = librosa.feature.mfcc(y, sr=sr)
            MFCC_blues.append(mfcc_blues)

            # Spectral Centroid 
            spectral_centroids_blues = librosa.feature.spectral_centroid(y, sr=sr)
            CENTROID_blues.append(spectral_centroids_blues)


            # Spectral Bandwidth
            spectral_bandwidth_blues = librosa.feature.spectral_bandwidth(y+0.01, sr=sr) #When  p=2 , this is like a weighted standard deviation.
            BANDWIDTH_blues.append(spectral_bandwidth_blues) 

            # Spectral Rolloff
            spectral_rolloff_blues = librosa.feature.spectral_rolloff(y+0.01, sr=sr)
            ROLLOFF_blues.append(spectral_rolloff_blues)
```

```
%%time

# calcolo media e varianza di ogni feature audio estratta
i= 0

count_blues = len(MFCC_blues)

MEAN_ZCR_blues= []
VAR_ZCR_blues = []

MEAN_HARMONIC_blues= []
VAR_HARMONIC_blues = []

MEAN_PERCUSSIVE_blues= []
VAR_PERCUSSIVE_blues = []

MEAN_MFCC_blues = []
VAR_MFCC_blues = []

MEAN_CHROMA_blues = []
VAR_CHROMA_blues = []

MEAN_RMSE_blues = []
VAR_RMSE_blues = []

MEAN_BANDWIDTH_blues = []
VAR_BANDWIDTH_blues = []

MEAN_CENTROID_blues = []
VAR_CENTROID_blues = []

MEAN_ROLLOFF_blues = []
VAR_ROLLOFF_blues = []


while i < count_blues:

  mean_zcr_blues = ZCR_blues[i].mean()
  MEAN_ZCR_blues.append(mean_zcr_blues)

  var_zcr_blues = ZCR_blues[i].var()
  VAR_ZCR_blues.append(var_zcr_blues)

  mean_harmonic_blues = HARMONIC_blues[i].mean()
  MEAN_HARMONIC_blues.append(mean_harmonic_blues)

  var_harmonic_blues = HARMONIC_blues[i].var()
  VAR_HARMONIC_blues.append(var_harmonic_blues)

  mean_percussive_blues = PERCUSSIVE_blues[i].mean()
  MEAN_PERCUSSIVE_blues.append(mean_percussive_blues)

  var_percussive_blues = PERCUSSIVE_blues[i].var()
  VAR_PERCUSSIVE_blues.append(var_percussive_blues)


  mean_mfcc_blues = MFCC_blues[i].mean(axis=1)
  MEAN_MFCC_blues.append(mean_mfcc_blues)

  var_mfcc_blues = MFCC_blues[i].var(axis=1)
  VAR_MFCC_blues.append(var_mfcc_blues)

  mean_chroma_blues = CHROMA_blues[i].mean()
  MEAN_CHROMA_blues.append(mean_chroma_blues)

  var_chroma_blues = CHROMA_blues[i].var()
  VAR_CHROMA_blues.append(var_chroma_blues)

  mean_rmse_blues = RMSE_blues[i].mean()
  MEAN_RMSE_blues.append(mean_rmse_blues)

  var_rmse_blues = RMSE_blues[i].var()
  VAR_RMSE_blues.append(var_rmse_blues)

  mean_centroid_blues = CENTROID_blues[i].mean()
  MEAN_CENTROID_blues.append(mean_centroid_blues)

  var_centroid_blues = CENTROID_blues[i].var()
  VAR_CENTROID_blues.append(var_centroid_blues)

  mean_bandwidth_blues = BANDWIDTH_blues[i].mean()
  MEAN_BANDWIDTH_blues.append(mean_bandwidth_blues)

  var_bandwidth_blues = BANDWIDTH_blues[i].var()
  VAR_BANDWIDTH_blues.append(var_bandwidth_blues)

  mean_rolloff_blues = ROLLOFF_blues[i].mean()
  MEAN_ROLLOFF_blues.append(mean_rolloff_blues)

  var_rolloff_blues = ROLLOFF_blues[i].var()
  VAR_ROLLOFF_blues.append(var_rolloff_blues)

  i = i +1
```
